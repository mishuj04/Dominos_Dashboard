```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv("pizza_sales.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['pizza_ingredients'] = df['pizza_ingredients'].str.split(',')
```

```python
df['pizza_ingredients'] = df['pizza_ingredients'].apply(lambda x: [i.strip() for i in x])
```

```python
df.head()
```

```python
df["order_time"]=pd.to_datetime(df["order_time"], format='%H:%M:%S').dt.time
```

```python
df["order_hour"]=pd.to_datetime(df["order_time"],format='%H:%M:%S').dt.hour
```

```python
def get_meal(hour):
    if 6<=hour<11:
        return "Breakfast"
    elif 11<=hour<16:
        return "Lunch"
    elif 16<=hour<22:
        return "Dinner"
    else:
        return "Late Night"
df["meal_time"]=df["order_hour"].apply(get_meal)
```

```python
df.head()
```

```python
df["order_date"]=pd.to_datetime(df["order_date"])
```

```python
df["weekday_number"]=df["order_date"].dt.weekday
```

```python
df.head()
```

```python
def weekend(day):
    if day==5 or day==6:
        return 1
    else:
        return 0
df["weekend"]=df["weekday_number"].apply(weekend)
```

```python
df.head()
```

```python
df["ingredient_count"]=len(df["pizza_ingredients"])
```

```python
df.to_csv('output.csv', index=False)
```
